##  1. Given the following implementation of the class `PriorityQueue`, implement the methods:

- `insert(v,k)` - add an element $v$ with priority $k$. Complexity $O(n)$
- `deleteMin()` - remove the element with the lowest $k$ (highest priority). Complexity $O(1)$
- `decreaseKey(v,k)` - decrease the value of $k$ (increase priority). Complexity $O(n)$

**Show complexity analysis for each implementation**

```python
import ctypes
```

```python
class PriorityQueue(object):

    def __init__(self, n):
        self.item_count = 0
        self.n = n
        self.queue = self._create_queue(self.n)

    def _create_queue(self, n):
        return (n * ctypes.py_object)()

    def enqueue(self, item):
        if self.item_count == self.n:
            raise ValueError("No más capacidad")

        self.queue[self.item_count] = item
        self.item_count += 1

    def enqueueOrder(self, item):
        if self.item_count == self.n:
            raise ValueError("No más capacidad")

        self.queue[self.item_count] = item
        self.item_count += 1

        for j in range(1, self.item_count):
            key = self.queue[j]
            i = j - 1
            while i > -1 and self.queue[i] >= key:
                self.queue[i + 1] = self.queue[i]
                i = i - 1
            self.queue[i + 1] = key

    def dequeue(self):
        c = self.queue[0]
        for i in range(1, self.item_count):
            self.queue[i - 1] = self.queue[i]
        self.queue[self.item_count - 1] = ctypes.py_object
        self.item_count -= 1
        return c

    def first(self):
        return self.queue[0]

    def full(self):
        if self.item_count == self.n:
            return True
        else:
            return False

    def empty(self):
        if self.item_count == 0:
            return True
        return False

    def size(self):
        return self.item_count

    def decreaseKey(self, item):
        for i in range(self.item_count):
            if self.queue[i] == item:
                buscado = self.queue[i]
                if buscado[0] == 0:
                    print("No se puede priorizar más")
                    return
                key = buscado[0]-1
                valor = buscado[1]
                itemNuevo = (key, valor)
                self.queue[i] = itemNuevo
                return
```

## 2. You are given two non-negative integers in the form of two non-empty linked lists. The digits are stored in reverse order, and each nodes contains a single digit. Add the two numbers and return the sum as a linked list.

For example:

**Input:** 
- $L_1$ = 1 -> 4 -> 5
- $L_2$ = 4 -> 3 -> 2

**Output:**
- $L_3$ = 5 -> 7 -> 7

Note that, the problem is equivalent to adding: 541 + 234  =  775

```python
class Node:
    def __init__(self, val=None):
        self.val = val
        self.next_node = None

    def set_next_node(self, next_node):
        self.next_node = next_node
```

```python
class SinglyLinkedList:
    # MÉTODO INICIALIZADOR
    def __init__(self, head_node=None):
        self.head_node = head_node

    # MÉTODO PARA IMPRIMIR TODOS LOS NODOS DE LA LISTA
    def list_traverse(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node

    # MÉTODO PARA AÑADIR UN NODO AL INICIO
    def insert_beginning(self, new_head_node):
        old_head_node = self.head_node
        self.head_node = new_head_node
        new_head_node.set_next_node(old_head_node)

    def insert_end(self, new_node):
        if self.head_node is None:
            self.head_node = new_node
            return

        node = self.head_node
        while node.next_node is not None:
            node = node.next_node

        node.set_next_node(new_node)

    def insert_given(self, new_node, given_node):
        node = self.head_node
        node_next = self.head_node
        while node:
            node = node.next_node
            node_next = node_next.next_node
            if node is given_node:
                node_next = node_next.next_node
                break

        node.set_next_node(new_node)
        new_node.set_next_node(node_next)

    def length(self):
        n = 0
        node = self.head_node
        while node:
            node = node.next_node
            n += 1
        return n
```

```python
numero1 = input('Inserta el primer número: ')
numero2 = input('Inserta el segundo número: ')


listaN1 = SinglyLinkedList()
listaN2 = SinglyLinkedList()
listaSuma = SinglyLinkedList()

for digit in numero1:
    nodo = Node(digit)
    listaN1.insert_beginning(nodo)

for digit in numero2:
    nodo = Node(digit)
    listaN2.insert_beginning(nodo)

longitud = listaN1.length()
# VOY A ASUMIR QUE LAS LONGITUDES DE AMBOS NÚMEROS ES LA MISMA length1 = length2

nodoL1 = listaN1.head_node
nodoL2 = listaN2.head_node
exceso = 0

for n in range(longitud):
    digitoN1 = int(nodoL1.val)
    digitoN2 = int(nodoL2.val)
    digitoSuma = digitoN1 + digitoN2
    if digitoSuma > 9:
        exceso = 1
        segundoDigito = str(digitoSuma)[1]
        digitoSuma = int(segundoDigito)

    nodoL1 = nodoL1.next_node
    nodoL2 = nodoL2.next_node

    if n > 0:
        digitoSuma = digitoSuma + exceso

    nodo = Node(digitoSuma)
    listaSuma.insert_end(nodo)

    if n == longitud-1:
        nodoExtra = Node(exceso)
        listaSuma.insert_end(nodoExtra)


listaSuma.list_traverse()
```

**Output:**

```python
    def detectLoop(self):
        tortoise = self.head_node
        hare = self.head_node
        contador = 0
        while tortoise and hare and hare.next_node:
            tortoise = tortoise.next_node
            hare = hare.next_node.next_node
            contador += 1
            if tortoise == hare:
                print("El bucle inicia en el nodo: ", tortoise.val)
                print("Se encuentra en el nodo numero: ", contador)
                return True

```

```python
class Deque():

    #Los números a continuación son los index de los elementos que después se van a estudiar
    def __init__(self, n):
        self.n = n
        self.item_count = 0
        self.head = 0
        self.tail = n-1
        self.array = [0]*n

    def enqueueFirst(self, item):
        self.array[self.head] = item
        self.item_count += 1
        self.head = (self.head+1)%self.n

    def enqueueLast(self, item):
        self.array[self.tail] = item
        self.item_count += 1
        self.tail = (self.tail-1)%self.n

    def dequeueFirst(self):
        self.item_count -= 1
        self.head = (self.head-1)%self.n
        c = self.array[self.head]
        return c

    def dequeueLast(self):
        self.item_count -= 1
        self.tail = (self.tail+1)%self.n
        c = self.array[self.tail]
        return c

    def printFirstLast(self):
        print(self.head, "  -   ", self.tail)
        print(self.array)
```

```python
pruebaDeque = Deque(10)
pruebaDeque.enqueueFirst(2)
pruebaDeque.enqueueFirst(3)
pruebaDeque.enqueueFirst(4)
pruebaDeque.enqueueLast(5)
pruebaDeque.enqueueLast(6)
pruebaDeque.enqueueLast(6)
pruebaDeque.printFirstLast()
print(pruebaDeque.dequeueLast())
print(pruebaDeque.dequeueFirst())

pruebaDeque.printFirstLast()
```

```python
class Stack(object):
    def __init__(self, n):
        self.item_count = 0
        self.n = n
        self.stack = self._create_stack(self.n)

    def _create_stack(self, n):
        return (n * ctypes.py_object)()

    def push(self, item):
        if self.item_count == self.n:
            raise ValueError("no more capacity")
        self.stack[self.item_count] = item
        self.item_count += 1

    def pop(self):
        c = self.stack[self.item_count - 1]
        self.stack[self.item_count] = ctypes.py_object
        self.item_count -= 1
        return c

    def top(self):

        return self.stack[self.item_count - 1]

    def full(self):

        if self.item_count == self.n:
            return True
        return False

    def empty(self):
        if self.item_count == 0:
            return True
        return False

    def size(self):
        return self.item_count

```

```python
class StackQueue():
    def __init__(self, n):
        self.stack1 = Stack(n)
        self.stack2 = Stack(n)

    def enqueue(self, item):
        while self.stack1.size() != 0:
            self.stack2.push(self.stack1.top())
            self.stack1.pop()

        self.stack1.push(item)

        while self.stack2.size() != 0:
            self.stack1.push(self.stack2.top())
            self.stack2.pop()

    def dequeue(self):
        if self.stack1.size() == 0:
            print("Está vacío")

        item = self.stack1.top()
        self.stack1.pop()
        return item
```

```python
pruebaStackQueue = StackQueue(10)
pruebaStackQueue.enqueue("Primer elemento ingresado")
pruebaStackQueue.enqueue("Segundo elemento ingresado")
pruebaStackQueue.enqueue("Tercer elemento ingresado")

print(pruebaStackQueue.dequeue())
print(pruebaStackQueue.dequeue())
print(pruebaStackQueue.dequeue())
```

```python
class Queue(object):
    def __init__(self, n):
        self.item_count = 0
        self.n = n
        self.queue = self._create_queue(self.n)

    def _create_queue(self, n):
        return (n * ctypes.py_object)()

    def enqueue(self, item):
        if self.item_count == self.n:
            raise ValueError("No más capacidad")
        self.queue[self.item_count] = item
        self.item_count += 1

    def dequeue(self):
        c = self.queue[0]
        for i in range(1, self.item_count):
            self.queue[i - 1] = self.queue[i]
        self.queue[self.item_count - 1] = ctypes.py_object
        self.item_count -= 1
        return c

    def first(self):
        return self.queue[0]

    def full(self):
        if self.item_count == self.n:
            return True
        else:
            return False

    def empty(self):
        if self.item_count == 0:
            return True
        return False

    def size(self):
        return self.item_count

```

```python
class QueuesStack():
    def __init__(self, n):
        self.queue1 = Queue(n)
        self.queue2 = Queue(n)
        self.queue = Queue(n)
        self.item_count = 0

    def push(self, item):
        self.item_count += 1
        self.queue2.enqueue(item)

        while not self.queue1.empty():
            self.queue2.enqueue(self.queue1.first())
            self.queue1.dequeue()

        self.queue = self.queue1
        self.queue1 = self.queue2
        self.queue2 = self.queue

    def pop(self):
        if self.queue1.empty():
            return
        self.queue1.dequeue()
        self.item_count -= 1

    def top(self):
        if self.queue1.empty():
            return -1
        return self.queue1.first()

    def size(self):
        return self.item_count
```

```python
pruebaStack = QueuesStack(10)

pruebaStack.push(1)
pruebaStack.push(2)
pruebaStack.push(3)

print(pruebaStack.top())
pruebaStack.pop()
print(pruebaStack.top())
pruebaStack.pop()
print(pruebaStack.top())
```
