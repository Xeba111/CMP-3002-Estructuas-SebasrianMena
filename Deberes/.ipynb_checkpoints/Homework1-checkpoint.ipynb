## 2. Implement in Python the insertion sort procedure to sort into non-increasing instead of non-decreasing order 

1. Use the `time` function to measure the execution time for the best and worst inputs of size between 10 and 1,000 (use steps of 10)
2. Plot the best and worst execution times measured in (2.1) as a function of $n$
3. Use the `random` function to generate randomly sorted inputs to calculate the execution time. For each $n$ run the program for 100 different inputs. Do for $n = 100,200, \dots, 1000$.

<b> Ejercicio 2-1 y 2-2</b>

```python

def InverseInsertionSort(array):
    for i in range(1, len(array)):        
        key = array[i]
        j = i-1
        while (j >= 0) and (key > array[j]):
            array[j+1] = array[j]
            j -= 1
        array[j+1] = key
```

```python
def CreateArrayWorst(num):
    array = []
    for i in range(1, num):
        array.append(i)
    return array
```

```python
def TimeWorstCase(num):
    array = CreateArrayWorst(num)
    start = time.time()
    InverseInsertionSort(array)
    return time.time()-start
```

```python
f1 = pd.DataFrame()
f1['n'] = range(10, 1000, 10)
f1['PC'] = [TimeWorstCase(x) for x in f1['n']]
f1['PC'] = 1000*f1['PC']
```

```python
f1.plot(x='n', y='PC')
plt.ylabel('Milisegundos')
plt.grid()
```

```python
def CreateArrayBest(num):
    array = []
    for i in range(num, 0, -1):
        array.append(i)
    return array
```

```python
def TimeBestCase(num):
    array = CreateArrayBest(num)
    start = time.time()
    InverseInsertionSort(array)
    return time.time()-start
```

```python
f2 = pd.DataFrame()
f2['n'] = range(10, 1000, 10)
f2['MC'] = [TimeBestCase(x) for x in f2['n']]
f2['MC'] = 1000*f2['MC']
```

```python
f2.plot(x='n', y='MC')
plt.ylabel('Milisegundos')
plt.grid()
```

Muestre que para cualquier constante real $a,b$, donde $b>0:$
<br><br>
$(n+a)^b = \Theta(n^b)$

Hagamos que $c=2^b$ y $n_0 \geq 2a$
<br><br>
Entonces, para cada $n\geq n_0$ tenemos que $(n+a)^b \leq (2n)^b = cn^b$ por lo que $(n+a)^b = O(n^b)$
<br><br>
Con esto podemos decir que $n_0 \geq \frac{-a}{1-(1/2)^{1/b}}$ y $c=0.5$. 
<br><br>
Entonces, $n \geq n_0 \geq n_0 \geq \frac{-a}{1-(1/2)^{1/b}}$ si y solo si $n-\frac{n}{2^{1/b}} \geq -a$, si y solo si $n+a \geq (1/2)^{a/b}n$ y si y solo si $(n+a)^b \geq cn^b$. Por lo tanto, $(n+a)^b = \Omega(n^b)$. Aplicamos el teorema 3.1 y queda demostrado todo.
